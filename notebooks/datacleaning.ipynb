```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(r"C:\Users\Reference\Downloads\raw data\dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["square_feet"] = df["square_feet"].str.extract('(\d+)')
df["square_feet"] = df["square_feet"].astype(float)
```

```python
df["price_per_sqft"] = (
    df["price_per_sqft"]
    .str.replace("₹","", regex=False)
    .str.replace(",","", regex=False)
    .str.extract('(\d+)')
    .astype(float)
)
```

```python
df["price"] = df["price"].astype(str)

df["price"] = df["price"].str.replace("₹","", regex=False)
df["price"] = df["price"].str.replace("Lac","", regex=False)
df["price"] = df["price"].str.replace(",","", regex=False)

df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
df["transaction"] = df["transaction"].fillna("Unknown")
df["furnishing"] = df["furnishing"].fillna("Unknown")
df["facing"] = df["facing"].fillna("Unknown")
df["floor"] = df["floor"].fillna("Unknown")
df["status"] = df["status"].fillna("Unknown")
```

```python
df = df.drop(columns=["description"])
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#handling the remaining missing value of square_feet
df["square_feet"] = df["square_feet"].fillna(df["square_feet"].median())
```

```python
#handling the remaining missing value of price_per_sqft
df["price_per_sqft"] = df["price_per_sqft"].fillna(df["price_per_sqft"].median())
```

```python
#handling the remaining missing value of price
df["price"] = df["price"].fillna(df["price"].median())
```

```python
#checking for final cleaned data
df.isnull().sum()
```

```python
df["furnishing"].unique()
```

```python
df["furnishing"] = df["furnishing"].replace({
    "1": "Unknown",
    "2": "Unknown",
    "Anand Asp": "Unknown"
})
```

```python
df["facing"] = df["facing"].replace({
    "1": "Unknown",
    "2": "Unknown"
})
```

```python
df["floor_number"] = df["floor"].str.extract('(\d+)')
df["floor_number"] = pd.to_numeric(df["floor_number"], errors="coerce")
```

```python
df["floor_number"] = df["floor_number"].fillna(df["floor_number"].median())
```
